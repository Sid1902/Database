### STEP 6: Query or filter out data in MongoDB

````python
for i in faculties.find({'Faculty':"Virat Sagar"}):
    print(i)
````

````python
for i in faculties.find({'_id':{'$gt':'1'}}):
    print(i)
````

````python
query1 = {"_id": '1'}

results = faculties.find(query1)
for data in results:
    print(data)
````

````python
query2 = {"_id": {"$gt": "1"}}

results = faculties.find(query2)
for data in results:
    print(data)
````
